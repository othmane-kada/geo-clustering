```python
import pandas as pd 
db = pd.read_csv('/home/othmanekada/data/sim.csv',skiprows=1,names=['time','provider' ,'accuracy', 'timestamp' ,'deviceid' ,'idtype' ,'altitude' ,'latitude' ,'longitude','','','','','y','m','d','h'])
#db=db.drop_duplicates(['longitude','latitude'])
db['latitude']=db['latitude']/1000000
db['longitude']=db['longitude']/1000000
dt=db[['accuracy','latitude' ,'longitude','y','m','d','h']]
days=dt.drop_duplicates(['m','y','d','h'])
dt
```

```python
import scipy.cluster.hierarchy as hcluster
import numpy as np
#print("tmp",tmp)
for i, r in days.iterrows():
    tmp=dt[(dt['m']==r['m']) & (dt['y']==r['y'])& (dt['d']==r['d'])]
    tmp=tmp.drop_duplicates(['longitude','latitude','h','accuracy'])
    tmp.index = range(len(tmp.index))
    hours=tmp.drop_duplicates(['h'])
    hours.index = range(len(hours.index))
    #print(hours)
    for j, rj in hours.iterrows():
       #print(rj['h'])
        tc=tmp[(tmp['h']==rj['h'])]
        tc=tc.drop_duplicates()
        tc.index = range(len(tc.index))
        tc=tc[['latitude','longitude','accuracy']]
        kms = 6371
        #print(tc)
        if(tc.latitude.count() > 1):
                 clusters=hcluster.fclusterdata(np.radians(tc[['latitude','longitude']]),0.1/kms, criterion="distance")
                 tc['c']=clusters
tc        
```

```python
cnt=tc.drop_duplicates('c')
cnt.index = range(len(cnt.index))
for i in cnt.index:
    cnt.at[i,'latitude']=tc[tc['c']==i+1].latitude.mean()
    cnt.at[i,'longitude']=tc[tc['c']==i+1].longitude.mean()

print("originale",tc.c.count())
print("clustered",cnt.c.count())
tc.accuracy.min()
```
